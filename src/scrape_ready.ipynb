# Imports and definitions


```python
# coding: utf-8
# For scraping/crawling
import csv
import urllib3
import certifi
from bs4 import BeautifulSoup


ERROR_MESSAGES = []
ERROR_FILE = "errors.txt"
WARNINGS = []
WARNINGS_FILE = "warnings.txt"
HEADERS = ["title", "data", "link"]

# Link relevant information
BASE_PAGE = "https://www.icmc.usp.br/eventos"
```

# What is a Balburdia?


```python
class Balburdia(object):
    """
    The Appliance class holds all usefull information for a appliance.
    Some of this information is pertinent to the customer, some of it is pertinent for our IA model.
    """
    def __init__(self, title, data, link):
        self.title = title
        self.data = data
        self.link = link

    def get_balburdia_info(self):
        return [self.title, self.data, self.link]

    def __str__(self):
        return ", ".join(self.get_balburdia_info())
```

# Connect to URL


```python
def connect(url):
    http = urllib3.PoolManager(cert_reqs='CERT_REQUIRED', ca_certs=certifi.where())
    headers = {}
    headers["User-Agent"] = "Mozilla/5.0 (X11; Linux x86_64)\
            AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
    headers["X-Requested-With"] = "XMLHttpRequest"

    # Try to get the page and the soup.
    # A better implementation would use try/except to catch errors.
    req = http.request(
        "GET",
        url,
        headers=headers
    )

    page = req.data
    return BeautifulSoup(page, "lxml")
```

# Helper functions


```python
def save_balburdias(file_name, balburdia_list):
    # JSON
    with open(file_name, "w", newline="") as csvfile:
        content_writer = csv.writer(csvfile)
        content_writer.writerow(HEADERS)
        for item in balburdia_list:
            content_writer.writerow([str(x) for x in item.get_balburdia_info()])


def scrape_balburdias(page):
    soup = connect(page)
    items = []
    
    for item in soup.find_all("div", {"class": "col-xs-12 col-sm-6 col-lg-3 quadro"}):
        link = item.find("a", href=True)["href"]
        if link[0] == "/":
            continue
            
        data = item.find("p").text
        title = item.find("h4").text
        print(f"{title} - {data} - {link}", "\n\n")
        items.append(Balburdia(title, data, link))

    return items

            
```

# Main


```python
def main():
    balburdias = scrape_balburdias(BASE_PAGE)
    save_balburdias("balburdias.csv", balburdias)

    print(f"{len(balburdias)} balburdas were scraped!")

    print("{} warnings were found. View them in the {} file.".format(
          len(WARNINGS), WARNINGS_FILE))
    print("{} errors were found. You can also view them in the {} file.".format(
        len(ERROR_MESSAGES), ERROR_FILE))
    print(*ERROR_MESSAGES, sep="\n")

    with open(ERROR_FILE, "w") as outfile:
        outfile.write("\n".join(ERROR_MESSAGES))

    with open(WARNINGS_FILE, "w") as outfile:
        outfile.write("\n".join(WARNINGS))

if __name__ == "__main__":
    main()
```
